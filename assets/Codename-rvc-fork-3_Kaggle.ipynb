## Install
Below in the code you can change the `branch` string value, The branch you want to use, for example `main` which is the default one and gets earlier but unstable updates, or release (which you can get by going to one of the releases which might be more stable, then clicking the tag and just paste the name like `v3.1.6-rev2`)

```python
import codecs
from IPython.display import clear_output
rot_47 = lambda encoded_text: "".join(
    [
        (
            chr(
                (ord(c) - (ord("a") if c.islower() else ord("A")) - 47) % 26
                + (ord("a") if c.islower() else ord("A"))
            )
            if c.isalpha()
            else c
        )
        for c in encoded_text
    ]
)

new_name = rot_47("kmjbmvh_hg")
findme = rot_47(codecs.decode("pbbxa://oqbpcj.kwu/Dqlitvb/qurwg-mtnqvlmz.oqb", "rot_13"))
uioawhd = rot_47(codecs.decode("pbbxa://oqbpcj.kwu/kwlmvium0wo/kwlmvium-zdk-nwzs-3", "rot_13"))

branch = "main"

!pip install uv
!git clone --depth 1 $uioawhd $new_name --branch $branch
clear_output()
!mkdir -p /kaggle/tmp
%cd /kaggle/tmp
!uv pip install torch==2.7.0 torchvision torchaudio==2.7.0 --upgrade --index-url https://download.pytorch.org/whl/cu128 -q
!uv pip install -r /kaggle/working/program_ml/requirements.txt -q
%cd /kaggle/working/program_ml
!python core.py "prerequisites" --models "True" --exe "True" --pretraineds_hifigan "True" > /dev/null 2>&1
!sudo curl -fsSL https://raw.githubusercontent.com/filebrowser/get/master/get.sh | sudo bash
!filebrowser config init
!filebrowser config set --auth.method=noauth
!filebrowser users add  "codename" "" --perm.admin
clear_output()
print("Finished")
```
